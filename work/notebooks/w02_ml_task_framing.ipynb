## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

```python
'Ranking'
```

I chose ranking because my goal is to prioritize webpages based on multiple ranking signals. Instead of simply labeling pages as good or bad, I want to order them from highest priority to lowest priority for SEO improvement. This helps SEO teams decide which pages to work on first when time and resources are limited. Ranking uses several signals together, making it more useful than a simple fixed rule based on only one feature.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
'Target/Proxy: My target is an SEO priority ranking for each webpage. Since the dataset does not contain a direct column showing which pages should be optimized first, I will use a proxy based on ranking signals such as average position, CTR, impressions, engagement rate, and content age. This is a defined proxy rather than an observed outcome because the priority is inferred from existing features instead of being directly recorded in the dataset.'
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
'Precision@K'
```

Success Metric: I will use Precision@K as my success metric. This measures how many of the top-ranked pages recommended by the model are actually relevant for SEO action. Precision@K is appropriate because SEO teams usually have limited time and resources, so they focus on improving only the highest-priority pages. A higher Precision@K indicates that the model is placing more relevant pages near the top of the ranking, making its recommendations more useful.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed rule can only check one or two conditions, such as "if CTR is below 5%, then improve the page." However, webpage performance depends on many ranking signals working together, including average position, impressions, engagement rate, content age, and freshness. Machine learning can analyze these signals together and identify more complex patterns than a simple if-statement. This helps SEO teams make better decisions about which pages should be prioritized.
